```r
# This script loads a RGB image and plots its L channel

library(ggplot2)

#Sets the working directory
#setwd("E:/CHB/PASANTIA/imagenes_forestal")

#Loads the imager library
library(imager)
```

El primer paso consiste en cargar la imagen en una variable.

```r
#Load and plot the RGB file
#data.file <- '/home/zamu/software/imagenes_forestal/data/9A9.jpg'
#data.file <- '/mnt/usb-WD_Elements_25A2_575852314531383859503438-0:0-part1/CHB/PASANTIA/imagenes_forestal/data/DJI_0805.jpg'
data.file <- '../imagenes_forestal/data/DJI_0805.jpg'
im <- load.image(data.file)
```

Una vez que la imagen se encuentra cargada en la variable, es posible imprimir el gráfico con la función <code>plot</code>

```r
plot(im)
```

El trabajo requiere que las imágenes sean tratadas desde la perspectiva **L** del [HSL](https://es.wikipedia.org/wiki/Modelo_de_color_HSL).

```r
#Gets the HSL colorspace from image
im_hsl <- RGBtoHSL(im)
```

```r
plot(im_hsl)
```

```r
#prints the image L channel from the HSL colorspace. index 3 is L
plot(channel(im_hsl, 3))
```

```r
#Extracts the L component from HSL colorspace image
im_L <- im_hsl[,,3]
```
